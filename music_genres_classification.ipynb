```python
from transformers import pipeline

# Crée un pipeline de classification audio
classifier = pipeline(
    task="audio-classification",
    model="dima806/music_genres_classification",
    device=-1  # -1 = CPU, ou 0 si tu as un GPU
)
```

```python
audio_path = "AMBNaut_Goelands sur le port (ID 2573)_LS.mp3"  # ton fichier

preds = classifier(audio_path, top_k=5)

print("Prédictions :")
for p in preds:
    # p ressemble à {"label": "rock", "score": 0.92}
    print(f"{p['label']:10s} -> {p['score']:.3f}")
```
