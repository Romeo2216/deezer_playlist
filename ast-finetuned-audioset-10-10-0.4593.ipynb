```python
from transformers import pipeline

# Pipeline d’AudioSet avec AST (gros modèle, bien entraîné)
classifier = pipeline(
    task="audio-classification",
    model="MIT/ast-finetuned-audioset-10-10-0.4593",
    device=-1,      # CPU ; mets 0 si tu as un GPU CUDA
)

audio_path = "AMBNaut_Goelands sur le port (ID 2573)_LS.mp3"  # adapte à ton fichier

preds = classifier(audio_path, top_k=15)

print("Top tags AST :")
for p in preds:
    print(f"{p['label']:40s} -> {p['score']:.3f}")
```
